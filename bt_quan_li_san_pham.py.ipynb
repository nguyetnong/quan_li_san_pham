```python
#quan li san pham


class Sanpham():
    def __init__(self,ma_sp,tensp):
        self.ma_sp=ma_sp
        self.tensp=tensp
      
        
    def thongtinsp(self):
        print("ma_sp:" ,self.ma_sp)
        print("ten san pham:",self.tensp)
        
class list_sanpham():
    def __init__(self):
        self.list_sanpham=[]
        
    def them_sp(self):
        ma_san_pham=input("nhap ma_sp: ")
        ten_sp=input("nhap ten sp: ")
        sanpham=Sanpham(ma_sp=ma_san_pham,tensp=ten_sp)
        self.list_sanpham.append(sanpham)
        print("___________")
        print("vua them sp: ")
        sanpham.thongtinsp()
    def xem_all_sanpham(self):
        for sanpham in self.list_sanpham:
            sanpham.thongtinsp()
            
    def tim_theo_ma_sp(self):
        ma_san_pham=input("nhap id sp:")
        resulst=[]
        for i in self.list_sanpham:
            if i.ma_sp==ma_san_pham:
                resulst.append(i)
        if len(resulst)<=0:
            print("ma sp khong ton tai")
        else:
            for sanpham in resulst:
                sanpham.thongtinsp()       
                
    def tim_theo_tensp(self):
        ten_san_pham=input("nhap ten sp:")
        resulst=[]
        for sanpham in self.list_sanpham:
            if sanpham.tensp==ten_san_pham:
                resulst.append(sanpham)
        if len(resulst)<=0:
            print("ten sp khong ton tai")
        else:
            for sanpham in resulst:
                sanpham.thongtinsp()
            
    def update_sanpham(self):
        sanpham_up=input("nhap ma sp can update:  ")
       
        idx=0
        for sanpham in self.list_sanpham:
            if sanpham.ma_sp==sanpham_up:
                masp_new=input("nhap ma san pham new : ")
                self.list_sanpham[idx].ma_sp=masp_new
                ten_sanpham=input("nhap ten sp new : ")
                self.list_sanpham[idx].tensp=ten_sanpham
            else:
                idx+=1

    def xoa1_sanpham(self):#xoa sp theo ma san pham
        ma_san_pham=input("nhap ma san pham can xoa: ")
        for sanpham in self.list_sanpham:
            if sanpham.ma_sp==ma_san_pham:
                self.list_sanpham.remove(sanpham)
            
                
san_pham=list_sanpham()
while True:
    option=input("nhap hoat dong cua ban:")
    if option=="them":
        san_pham.them_sp()
    elif option == "xem":
        san_pham.xem_all_sanpham()
    elif option == "xemten":
        san_pham.tim_theo_tensp()
    elif option =="xemid":
        san_pham.tim_theo_ma_sp()
    elif option =="update":
        san_pham.update_sanpham() 
    elif option == "xoa":
        san_pham.xoa1_sanpham()
        
    print("_______________________________")
    
        
    
            
    
    
    
    
    
    
    
    
    
    
        

```
